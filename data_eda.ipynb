### puan dağılımı

```python
# Plotting side-by-side histograms
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Histogram for anksiyete1
axs[0].hist(anksiyete1['toplam puan'], bins=20, edgecolor='black')
axs[0].set_title('Anksiyete 1 Toplam Puan Dağılımı')

# Histogram for anksiyete2
axs[1].hist(anksiyete2['toplam puan'], bins=20, edgecolor='black')
axs[1].set_title('Anksiyete 2 Toplam Puan Dağılımı')

# Show plot
plt.tight_layout()
plt.show()
```

### yaş dağılımı

```python
# Plotting side-by-side histograms
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Histogram for anksiyete1
axs[0].hist(anksiyete1['Yaş'], bins=20, edgecolor='black')
axs[0].set_title('Anksiyete 1 Yaş Dağılımı')

# Histogram for anksiyete2
axs[1].hist(anksiyete2['Yaş'], bins=20, edgecolor='black')
axs[1].set_title('Anksiyete 2 Yaş Dağılımı')

# Show plot
plt.tight_layout()
plt.show()
```

### cinsiyet ve çalışma durumu

```python
# Data for anksiyete1 and anksiyete2 (assuming these are DataFrames or Series with 'cinsiyet'/'Cinsiyet' columns)
anksiyete1_gender_counts = anksiyete1['cinsiyet'].value_counts()
anksiyete2_gender_counts = anksiyete2['Cinsiyet'].value_counts()

# Plotting side-by-side pie charts
fig, axs = plt.subplots(1, 2, figsize=(10, 8))

# Pie chart for anksiyete1
axs[0].pie(anksiyete1_gender_counts, labels=anksiyete1_gender_counts.index, autopct='%1.1f%%', startangle=90)
axs[0].set_title('Anksiyete1 Cinsiyet Dağılımı')

# Pie chart for anksiyete2
axs[1].pie(anksiyete2_gender_counts, labels=anksiyete2_gender_counts.index, autopct='%1.1f%%', startangle=90)
axs[1].set_title('Anksiyete2 Cinsiyet Dağılımı')

plt.show()
```

```python
# Data for anksiyete1 and anksiyete2 (assuming these are DataFrames or Series with 'çalışma durumu' columns)
anksiyete1_employment_counts = anksiyete1['çalışma durumu'].value_counts()
anksiyete2_employment_counts = anksiyete2['çalışma durumu'].value_counts()

# Plotting side-by-side pie charts
fig, axs = plt.subplots(1, 2, figsize=(10, 8))

# Pie chart for anksiyete1
axs[0].pie(anksiyete1_employment_counts, labels=anksiyete1_employment_counts.index, autopct='%1.1f%%', startangle=90)
axs[0].set_title('Anksiyete1 Çalışma Durumu Dağılımı')

# Pie chart for anksiyete2
axs[1].pie(anksiyete2_employment_counts, labels=anksiyete2_employment_counts.index, autopct='%1.1f%%', startangle=90)
axs[1].set_title('Anksiyete2 Çalışma Durumu Dağılımı')

plt.show()
```
